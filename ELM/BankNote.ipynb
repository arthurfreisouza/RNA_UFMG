```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_banknote = pd.read_table('data_banknote_authentication.txt', sep = ',')
df_banknote
```

```python
df_banknote.shape
```

```python
df_banknote.rename(columns = {'3.6216' : 'variance', '8.6661' : 'skewness', '-2.8073' : 'curtosis', '-0.44699' : 'entropy', '0' : 'class'}, inplace = True)
df_banknote['class'] = df_banknote['class'].replace({0: -1})
df_banknote = df_banknote.astype(float)
df_banknote
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_banknote.iloc[ :, 0 : 4], df_banknote.iloc[ :, 4], random_state = 0, train_size = 0.7)
y_test[y_test == 0] = -1
```

```python
import train_test_ELM
p = 100
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
train_ELM = train_test_ELM.train_ELM(X_train, y_train, p, control = True)
w = np.array(train_ELM[0])
H = np.array(train_ELM[1])
Z = np.array(train_ELM[2])
```

```python
y_hat = train_test_ELM.test_ELM(X_test, Z, w, True)
print(np.sum(y_test - y_hat)**2 / y_hat.shape[0])
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_test, y_hat)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```
