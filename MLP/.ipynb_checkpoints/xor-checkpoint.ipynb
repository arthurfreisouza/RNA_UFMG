```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import time
import matplotlib.animation
```

```python
def sech2(x):
    y = np.array(2 / (np.exp(x) + np.exp(-x))) * np.array(2 / (np.exp(x) + np.exp(-x)))
    return y
```

```python
x = np.array((0, 0, 0, 1, 1, 0, 1, 1)).reshape(4, 2)
y = np.array((-1, 1, 1, -1, 1, -1, -1, 1)).reshape(4, 2)
print(f" Inputs : \n{x}. \n Labels : \n{y}.")
```

```python
wb11 = np.random.random(1) - 0.5
wb12 = np.random.random(1) - 0.5

wipt11 = np.random.random(1) - 0.5
wipt12 = np.random.random(1) - 0.5

wipt21 = np.random.random(1) - 0.5
wipt22 = np.random.random(1) - 0.5

print(f"Bias : ({wb11}, {wb12}). \n First Input : ({wipt11}, {wipt12}). \n Second Input : ({wipt21}, {wipt21}).")
```

```python
np.random.seed(np.random.randint(0, 10000))
wb21 = np.random.random(1) - 0.5
wb22 = np.random.random(1) - 0.5

wipt11_2 = np.random.random(1) - 0.5
wipt12_2 = np.random.random(1) - 0.5

wipt21_2 = np.random.random(1) - 0.5
wipt22_2 = np.random.random(1) - 0.5

print(f"Bias : ({wb21}, {wb22}). \n First Last Neuron Input : ({wipt11_2}, {wipt12_2}). \n Second Last Neuron Input : ({wipt21_2}, {wipt22_2}).")
```

```python
num_epochs = 0
tol = 0.1
err_epochs = tol + 1
max_epochs = 100
N = 4 # NÃºmero de entradas.
eta = 0.01 # Learning_rate.
err_vec = np.zeros(max_epochs)
```

```python
matH = np.zeros((N, 2))
```

```python
sequence = np.arange(start = 0, stop = N, step = 1).tolist()
seq_randomized = random.sample(population = sequence, k = N)
seq_randomized
out_n1 = np.zeros((max_epochs * N)).reshape(-1, N)
out_n2 = np.zeros((max_epochs * N)).reshape(-1, N)
```

```python
while((num_epochs < max_epochs) and (err_epochs > tol)):
    sequence = np.arange(start = 0, stop = N, step = 1).tolist()
    seq_randomized = random.sample(population = sequence, k = N)
    err2 = 0
    for i in range(N):
        # Pegando entradas aleatÃ³rias.
        rand_input = seq_randomized[i]
        input1 = x[rand_input, 0]
        input2 = x[rand_input, 1]


        # Pegando as saÃ­das das entradas aleatÃ³rias.
        output1 = y[rand_input, 0]
        output2 = y[rand_input, 1]

        # Inicializando o valor dos biases da camada de entrada e da camada de saÃ­da.
        bias1 = 1
        bias2 = 1


        # Fazendo a combinaÃ§Ã£o linear dos dois neurÃ´nios de saÃ­da e aplicando a funÃ§Ã£o de ativaÃ§Ã£o sigmoidal.
        cl1 = (bias1*wb11) + (input1*wipt11) + (input2*wipt21)
        cl2 = (bias1*wb12) + (input1*wipt12) + (input2*wipt22)
        tanh_n1 = np.tanh(cl1)
        tanh_n2 = np.tanh(cl2)



        # Fazendo a combinaÃ§Ã£o linear dos dois neutÃ´nios de entrada e aplicando a funÃ§Ã£o de ativaÃ§Ã£o sigmoidal.
        comb_n11 = (bias2*wb21) + (tanh_n1*wipt11_2) + (tanh_n2*wipt21_2)
        comb_n22 = (bias2*wb22) + (tanh_n1*wipt12_2) + (tanh_n2*wipt22_2)
        tanh_n11 = np.tanh(comb_n11)
        tanh_n22 = np.tanh(comb_n22)

        matH[rand_input, 0] = tanh_n11
        matH[rand_input, 1] = tanh_n22
        

        # Calculo do erro e a retropropagaÃ§Ã£o dos erros.
        err_output1 = (output1 - tanh_n11)
        err_output2 = (output2 - tanh_n22)
        delta_output1 = err_output1*sech2(comb_n11) # Lembrar que delta(N) = e*f'(u) SEMPRE.
        delta_output2 = err_output2*sech2(comb_n22) # Lembrar que delta(N) = e*f'(u) SEMPRE.



        # AtualizaÃ§Ã£o dos pesos : 
        dwbias21 = (eta*delta_output1*bias2) # atualizaÃ§Ã£o do peso do bias da camada de saÃ­da.
        dwbias22 = (eta*delta_output2*bias2) # atualizaÃ§Ã£o do peso do bias da camada de saÃ­da.

        
        dwipt_11_2 = (eta*delta_output1*tanh_n1) # atualizaÃ§Ã£o do peso do bias da camada de saÃ­da.
        dwipt_12_2 = (eta*delta_output2*tanh_n1) # atualizaÃ§Ã£o do peso do bias da camada de saÃ­da.


        dwipt_21_2 = (eta*delta_output1*tanh_n2) # atualizaÃ§Ã£o do peso do bias da camada de saÃ­da.
        dwipt_22_2 = (eta*delta_output2*tanh_n2) # atualizaÃ§Ã£o do peso do bias da camada de saÃ­da.
        
        # Calculando o delta dos neurÃ´nios da camada de entrada e a retropropragaÃ§Ã£o da camada de entrada.
        delta_input1 = (delta_output1*wipt11_2 + delta_output2*wipt12_2)*sech2(cl1)
        delta_input2 = (delta_output1*wipt21_2 + delta_output2*wipt22_2)*sech2(cl2)

        dbias11 = eta*delta_input1*bias1
        dbias12 = eta*delta_input2*bias1

        
        dipt11 = eta*delta_input1*input1
        dipt12 = eta*delta_input2*input1

        
        dipt21 = eta*delta_input1*input2
        dipt22 = eta*delta_input2*input2


        # AtualizaÃ§Ã£o de todos os pesos da rede.

        wb11 = wb11 + dbias11
        wb12 = wb12 + dbias12

        wipt11 = wipt11 + dipt11
        wipt12 = wipt12 + dipt12

        wipt21 = wipt21 + dipt21
        wipt22 = wipt22 + dipt22

        wb21 = wb21 + dwbias21
        wb22 = wb22 + dwbias22
        
        wipt11_2 = wipt11_2 + dwipt_11_2
        wipt12_2 = wipt12_2 + dwipt_12_2
        
        wipt21_2 = wipt21_2 + dwipt_21_2
        wipt22_2 = wipt22_2 + dwipt_22_2

        err2 = err2 + (err_output1**2 + err_output2)/4
    plt.scatter(matH[0:, 0], matH[0:, 1], color = 'blue')
    plt.scatter(matH[1:, 0], matH[1:, 1], color = 'red')
    plt.scatter(matH[2:, 0], matH[2:, 1], color = 'red')
    plt.scatter(matH[3:, 0], matH[3:, 1], color = 'blue')
    time.sleep(0.1)
    plt.show()
    err_vec[num_epochs] = err2/N
    num_epochs += 1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import time

# Dados iniciais
x = np.random.rand(10)
y = np.random.rand(10)

# Criar o grÃ¡fico inicial
sns.set(style="whitegrid")  # Estilo do grÃ¡fico
fig, ax = plt.subplots()
scatter = sns.scatterplot(x=x, y=y)

# ConfiguraÃ§Ãµes do grÃ¡fico
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.title("GrÃ¡fico VariÃ¡vel no Tempo")

# FunÃ§Ã£o para atualizar o grÃ¡fico
def update_plot():
    # Gerar novos dados
    x_new = np.random.rand(10)
    y_new = np.random.rand(10)
    
    # Limpar o grÃ¡fico
    ax.clear()
    
    # Plotar os novos pontos
    sns.scatterplot(x=x_new, y=y_new, ax=ax)
    
    # ConfiguraÃ§Ãµes do grÃ¡fico
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    plt.title("GrÃ¡fico VariÃ¡vel no Tempo")
    
    # ForÃ§ar a atualizaÃ§Ã£o do grÃ¡fico
    plt.draw()
    plt.pause(0.1)

# Loop para atualizar o grÃ¡fico
for i in range(10):
    update_plot()
    time.sleep(1)  # Pausa de 1 segundo para visualizaÃ§Ã£o

plt.show()
```
