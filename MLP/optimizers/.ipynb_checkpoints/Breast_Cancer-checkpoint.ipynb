```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
x = pd.DataFrame(data.data, columns = data.feature_names)
y = pd.Series(data.target)
x
```

```python
np.unique(y, return_counts = True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state =10)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
hidden_layer = (100, 10, 1)
clf = MLPClassifier(hidden_layer_sizes = hidden_layer , activation = 'tanh', solver = 'sgd', alpha = 0.01, max_iter = 500, learning_rate = 'adaptive', learning_rate_init = 0.01).fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
sum_train = sum(y_test != y_pred)
print(f"There are a total of {sum_train} errors in the train.")
```

```python
from sklearn.metrics import classification_report
y_pred_class_report = classification_report(y_test, y_pred, target_names=data.target_names)
print(y_pred_class_report)
```
