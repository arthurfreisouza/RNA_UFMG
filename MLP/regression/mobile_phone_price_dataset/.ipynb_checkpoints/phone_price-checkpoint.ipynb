```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('moc_phone_price.csv')
```

```python
df
```

```python
df.columns
```

```python
df['Price']=df['Price'].str.replace(',', '').astype(float) # Convert the 'Price' column to a numerical type
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(columns = ['Unnamed: 0', 'Name'], inplace = True)
df.drop_duplicates(inplace=True) # Remove duplicate rows if any
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoded_df = pd.get_dummies(df[['No_of_sim', 'Ram', 'company', 'Inbuilt_memory', 'fast_charging', 'Processor', 'Camera', 'Processor_name', 'Screen_resolution', 'Display', 'Battery', 'External_Memory', 'Android_version']], columns=['No_of_sim', 'Ram', 'company', 'Inbuilt_memory', 'fast_charging', 'Processor', 'Camera', 'Processor_name', 'Screen_resolution', 'Display', 'Battery', 'External_Memory', 'Android_version']).astype(int)
df.drop(columns=['No_of_sim', 'Ram', 'company', 'Inbuilt_memory', 'fast_charging', 'Processor', 'Camera', 'Processor_name', 'Screen_resolution', 'Display', 'Battery', 'External_Memory', 'Android_version'], inplace=True)
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

cond = ['Rating', 'Spec_score']
df.loc[:, cond] = scaler.fit_transform(df.loc[:, cond].astype(float))
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 5)
df = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)
```

```python
y = df.loc[ :, 'Price']
x = df.drop(columns = ['Price'])
```

```python
# Identify features with a correlation greater than 0.15 with the target variable
corr_u =df.corr()['Price'].sort_values(ascending=False, key=abs)>0.15 
corr_u.index = corr_u.index.str.replace(',','')
df.columns=df.columns.str.replace(',','')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.82, random_state = 45)
```

```python
%%time
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

k_val = np.arange(start = 1, stop = 100, step = 1)
neigh = KNeighborsRegressor()

regressor = GridSearchCV(estimator = neigh, param_grid = {'n_neighbors' : k_val}, cv = 10)
regressor.fit(X = X_train, y = y_train)

best_param = regressor.best_params_
best_score = regressor.best_score_
```

```python
print(f"The lowest MSE is using n_neighbor = {best_param['n_neighbors']}, and is : {best_score}")
```

```python
from sklearn.metrics import mean_squared_error

neigh_ = KNeighborsRegressor(n_neighbors = best_param['n_neighbors'])

neigh_.fit(X_train, y_train)
y_pred = neigh_.predict(X_test)

print(f" The MSE for the test values is {mean_squared_error(y_test, y_pred)}")
```

```python
results = pd.DataFrame({'y_real': y_test, 'y_pred': y_pred})
results
```

```python
%%time
from sklearn.neural_network import MLPRegressor

param_grid = {
    'hidden_layer_sizes': np.arange(start = 1, stop = 20, step = 1)
}


mlp = MLPRegressor(max_iter = 1000, early_stopping = True)

mlp_regressor = GridSearchCV(estimator = mlp, param_grid = param_grid, cv = 10)
mlp_regressor.fit(X = X_train, y = y_train)

best_mlp_score = mlp_regressor.best_score_
best_mlp_param = mlp_regressor.best_params_
```

```python
print(f"The lowest MSE is using hidden_layer_sizes = {best_mlp_param['hidden_layer_sizes']}, and is : {best_mlp_score}")
```

```python
mlp_reg = MLPRegressor(hidden_layer_sizes = best_mlp_param['hidden_layer_sizes'], max_iter = 1500)

mlp_reg.fit(X_train, y_train)
y_pred_mlp = mlp_reg.predict(X_test)

print(f" The MSE for the test values is {mean_squared_error(y_test, y_pred_mlp)}")
```

```python
results = pd.DataFrame({'y_real': y_test, 'y_pred': y_pred_mlp})
results
```
