#### Definindo a função secante

```python
def sech2(x):
    y = np.array(2 / (np.exp(x) + np.exp(-x))) * np.array(2 / (np.exp(x) + np.exp(-x)))
    return y
```

#### Iniciando os pesos aleatóriamente, é 1 entrada e 3 neurônios na camada intermediária.

```python
w14 = np.random.random(1) - 0.5
w15 = np.random.random(1) - 0.5
w16 = np.random.random(1) - 0.5

w24 = np.random.random(1) - 0.5
w25 = np.random.random(1) - 0.5
w26 = np.random.random(1) - 0.5

w37 = np.random.random(1) - 0.5
w47 = np.random.random(1) - 0.5
w57 = np.random.random(1) - 0.5
w67 = np.random.random(1) - 0.5

bias1 = 1
bias2 = 1
```

```python
num_epochs = 0
tol = 0.0001
err_epochs = tol + 1
max_epochs = 1000
N = x.shape[0] # Número de entradas.
eta = 0.01 # Learning_rate.
err_vec = np.zeros(max_epochs, dtype=float, order = 'C')
```

```python
x = np.linspace(start = 0, stop = 2*np.pi, num = 50)
y = np.sin(x)
```

```python
while((num_epochs < max_epochs) and (err_epochs > tol)):
    sequence = np.arange(start = 1, stop = N + 1, step = 1).tolist()
    seq_randomized = random.sample(population = sequence, k = N)
    err2 = 0
    for i in range(N):
        # Pegando entradas aleatórias.
        rand_input = seq_randomized[i] - 1
        input = x[rand_input]


        # Pegando as saídas das entradas aleatórias.
        output = y[rand_input]

        # Inicializando o valor dos biases da camada de entrada e da camada de saída.
        bias1 = 1
        bias2 = 1


        # Fazendo a combinação linear dos dois neurônios de saída e aplicando a função de ativação sigmoidal.
        comb_n1 = (bias1*w14) + (input*w24)
        comb_n2 = (bias1*w15) + (input*w25)
        comb_n3 = (bias1*w16) + (input*w26)
        tanh_n1 = np.tanh(comb_n1)
        tanh_n2 = np.tanh(comb_n2)
        tanh_n3 = np.tanh(comb_n3)


        # Fazendo a combinação linear dos dois neutônios de entrada e aplicando a função de ativação sigmoidal.
        output_net = (bias2*w37) + (tanh_n1*w47) + (tanh_n2*w57) + (tanh_n3*w67)
        
        # Calculo do erro e a retropropagação dos erros.
        err_output = (output - output_net)
        delta_output = err_output # Lembrar que delta(N) = e*f'(u) SEMPRE. Nesse caso, f'(u) = 1.



        # Atualização dos pesos : 
        dw37 = (eta*delta_output*bias2) # atualização do peso do bias da camada de saída.
        dw47 = (eta*delta_output*tanh_n1) # atualização do peso do bias da camada de saída.  
        dw57 = (eta*delta_output*tanh_n2) # atualização do peso do bias da camada de saída.
        dw67 = (eta*delta_output*tanh_n3) # atualização do peso do bias da camada de saída.


        
        # Calculando o delta dos neurônios da camada de entrada e a retropropragação da camada de entrada.
        delta_input1 = (delta_output*w47)*sech2(comb_n1)
        delta_input2 = (delta_output*w57)*sech2(comb_n2)
        delta_input3 = (delta_output*w67)*sech2(comb_n3)

        dw14 = eta*delta_input1*bias1
        dw15 = eta*delta_input2*bias1
        dw16 = eta*delta_input3*bias1

        dw24 = eta*delta_input1*input
        dw25 = eta*delta_input2*input
        dw26 = eta*delta_input3*input



        # Atualização de todos os pesos da rede.

        w14 = w14 + dw14
        w15 = w15 + dw15
        w16 = w16 + dw16

        
        w24 = w24 + dw24
        w25 = w25 + dw25
        w26 = w26 + dw26

        w37 = w37 + dw37
        w47 = w47 + dw47
        w57 = w57 + dw57
        w67 = w67 + dw67
        err2 = err2 + (err_output**2)
        #print(err2)

    err_vec[num_epochs] = err2/N
    err_epochs = err_vec[num_epochs]
    num_epochs += 1
```

```python
x_plot = np.linspace(start = 0, stop = num_epochs, num = num_epochs)
plt.plot(x_plot, err_vec, color = 'black', label = 'err_graphic')
plt.title('Err graphic')
plt.xlabel('num_epochs')
plt.ylabel('err²')
plt.legend()
plt.show()
```

```python
xrange = np.linspace(start = 0, stop = 2*np.pi, num = x.shape[0])
yhat = np.zeros((xrange.shape[0]))
N2 = xrange.shape[0]
```

```python
for i in range(N2):
    input2 = xrange[i]

    comb_n1_hidd = (bias1*w14) + (input2*w24)
    otp_n1_hidd   = np.tanh(comb_n1_hidd)

    comb_n2_hidd = (bias1*w15) + (input2*w25)
    otp_n2_hidd   = np.tanh(comb_n2_hidd)
    
    comb_n3_hidd = (bias1*w16) + (input2*w26)
    otp_n3_hidd   = np.tanh(comb_n3_hidd)

    final_otp = (bias2*w37) + (otp_n1_hidd*w47) + (otp_n2_hidd*w57) + (otp_n3_hidd*w67)

    yhat[i] = final_otp    
```
