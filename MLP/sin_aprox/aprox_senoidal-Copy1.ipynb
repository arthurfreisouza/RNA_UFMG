```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
import random
```

```python
def sech2(x):
    y = np.array(2 / (np.exp(x) + np.exp(-x))) * np.array(2 / (np.exp(x) + np.exp(-x)))
    return y
```

```python
x = np.random.uniform(low = 0, high = 2*np.pi, size = 100)
y = np.sin(x) + 0.1*np.random.normal(size = x.shape[0])
```

```python
lst_MSE = list()
lst_STD = list()
```

```python
plt.scatter(x, y, color = 'black', label = 'data generated')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('data random generated')
plt.legend
plt.show()
```

#### Inicializando os pesos da rede. A rede terá 1 camada intermediária com 3 neurônios e a saída de 2 neurônios

```python
wb11 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wb12 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wb13 = np.random.uniform(low = -0.5, high = 0.5, size = 1)

wipt11 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wipt12 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wipt13 = np.random.uniform(low = -0.5, high = 0.5, size = 1)

wb21 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wipt21 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wipt22 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wipt23 = np.random.uniform(low = -0.5, high = 0.5, size = 1) 
```

```python
bias1 = 1
bias2 = 1
```

```python
num_epochs = 0
tol = 0.0001
err_epochs = tol + 1
max_epochs = 10000
N = x.shape[0] # Número de entradas.
eta = 0.1 # Learning_rate.
err_vec = np.zeros(max_epochs, dtype=float, order = 'C')
```

```python
N = 10
sequence = np.arange(start = 0, stop = N, step = 1).tolist()
sequence
seq_randomized = random.sample(population = sequence, k = N - 1)
seq_randomized
```

```python
while((num_epochs < max_epochs) and (err_epochs > tol)):
    sequence = np.arange(start = 0, stop = N, step = 1).tolist()
    seq_randomized = random.sample(population = sequence, k = N)
    err2 = 0 # Erro de cada época.

    for i in range(N):
        rand_input = seq_randomized[i]
        
        ipt = x[rand_input] # Entrada aleatória
        opt = y[rand_input] # Saída aleatória


        
        # Fase feed forward.
        cl1 = (bias1*wb11) + (ipt*wipt11)
        act_cl1 = np.tanh(cl1)
        
        cl2 = (bias1*wb12) + (ipt*wipt12)
        act_cl2 = np.tanh(cl2)
        
        cl3 = (bias1*wb13) + (ipt*wipt13)
        act_cl3 = np.tanh(cl3)


        opt_net = (bias2*wb21) + (act_cl1*wipt21) + (act_cl2*wipt22) + (act_cl3*wipt23) # Saída da rede
        # opt_net = np.tanh(opt_net) para classificação.


        # Agora começa a parte da retropropagação do erro.

        err_opt = (opt - opt_net)
        delta_opt = err_opt # Lembrar que delta(N) = e*f'(u) SEMPRE. Nesse caso, f'(u) = 1.

        dbias2 = (eta*delta_opt*bias2)
        dipt21 = (eta*delta_opt*act_cl1)
        dipt22 = (eta*delta_opt*act_cl2)
        dipt23 = (eta*delta_opt*act_cl3)

        delta_ipt21 = (delta_opt*wipt21)*sech2(cl1)
        delta_ipt22 = (delta_opt*wipt22)*sech2(cl2)
        delta_ipt23 = (delta_opt*wipt23)*sech2(cl3)

        dbias11 = (eta*delta_ipt21*bias1)
        dbias12 = (eta*delta_ipt22*bias1)
        dbias13 = (eta*delta_ipt23*bias1)


        dipt11 = (eta*delta_ipt21*ipt)
        dipt12 = (eta*delta_ipt22*ipt)
        dipt13 = (eta*delta_ipt23*ipt)



        # Agora, com todas as equações de ajustes calculadas, basta atualizar cada peso da rede.

        # Atualizando os pesos do bias.
        wb11 = wb11 + dbias11
        wb12 = wb12 + dbias12
        wb13 = wb13 + dbias13

        
        # Atualizando pesos camada de entrada.
        wipt11 = wipt11 + dipt11
        wipt12 = wipt12 + dipt12
        wipt13 = wipt13 + dipt13


        # Atualizando pesos camada de saída.
        wb21 = wb21 + dbias2
        wipt21 = wipt21 + dipt21
        wipt22 = wipt22 + dipt22
        wipt23 = wipt23 + dipt23

        err2 = err2 + (err_opt**2)

    
    err_vec[num_epochs] = err2/N
    err_epochs = err_vec[num_epochs]
    num_epochs += 1
```

```python
x_plot = np.linspace(start = 0, stop = num_epochs, num = num_epochs)
plt.plot(x_plot, err_vec, color = 'black', label = 'err_graphic')
plt.title('Err graphic')
plt.xlabel('num_epochs')
plt.ylabel('err²')
plt.legend()
plt.show()
```

```python
xrange = np.linspace(start = 0, stop = 2*np.pi, num = x.shape[0])
y_true = np.sin(xrange)
yhat = np.zeros((xrange.shape[0]))
N2 = xrange.shape[0]
```

```python
for i in range(N2):
    ipt_test = xrange[i]

    cbl1 = (bias1*wb11) + (ipt_test*wipt11)
    otp_n1_hidd   = np.tanh(cbl1)

    cbl2 = (bias1*wb12) + (ipt_test*wipt12)
    otp_n2_hidd   = np.tanh(cbl2)
    
    cbl3 = (bias1*wb13) + (ipt_test*wipt13)
    otp_n3_hidd   = np.tanh(cbl3)

    final_otp = (bias2*wb21) + (otp_n1_hidd*wipt21) + (otp_n2_hidd*wipt22) + (otp_n3_hidd*wipt23)

    yhat[i] = final_otp    
```

```python
plt.plot(xrange, yhat, color = 'red', label = 'sin(x) aproxximation')
#plt.plot(xrange, np.sin(xrange) + 0.25, color = 'blue')
plt.plot(xrange, y_true, color = 'black', label = 'sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
lst_MSE.append(mean_squared_error(y_true, yhat))
lst_STD.append(statistics.stdev(mean_squared_error(y_true, yhat)))
```

```python
print(f"Using eta 0.1 and maxepochs 100, the model has the MSE : ({lst_MSE[0]} +- {lst_STD[0]})")
print(f"Using eta 0.01 and maxepochs 100, the model has the MSE : ({lst_MSE[1]} +- {lst_STD[1]})")
print(f"Using eta 0.01 and maxepochs 500, the model has the MSE : ({lst_MSE[2]} +- {lst_STD[2]})")
print(f"Using eta 0.01 and maxepochs 1000, the model has the MSE : ({lst_MSE[3]} +- {lst_STD[3]})")
print(f"Using eta 0.1 and maxepochs 1000, the model has the MSE : ({lst_MSE[4]} +- {lst_STD[4]})")
```
