```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.random.uniform(low = -10, high = 10, size = 500)
y = np.sin(x) + 0.3*np.random.normal(size = x.shape[0])
```

```python
plt.scatter(x = x, y = y, color = 'black')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 1, train_size = 0.8)
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
regr = MLPRegressor(hidden_layer_sizes=(500,10), random_state=1, max_iter=5000, activation = 'relu').fit(X_train, y_train)
y_pred = regr.predict(X_test)
```

```python
plt.scatter(X_test, y_pred, color = 'black')
plt.scatter(X_test, y_test,  color = 'red')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```
