```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 300
mean = 0
std_dev = 0.6

g1 = np.random.normal(mean, std_dev, 2*size).reshape(-1, 2) + [3, 4]
g2 = np.random.normal(mean, std_dev, 2*size).reshape(-1, 2) + [5, 2]
g3 = np.random.normal(mean, std_dev, 2*size).reshape(-1, 2) + [5, 4]
```

```python
plt.scatter(g1[:, 0], g1[:, 1], color= 'black', s = 2)
plt.scatter(g2[:, 0], g2[:, 1], color= 'black', s = 2)
plt.scatter(g3[:, 0], g3[:, 1], color= 'black', s = 2)
plt.title('Cluster')
plt.xlabel('x')
plt.ylabel('y')
```

```python
all_data = np.concatenate((g1, g2, g3), axis = 0)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state = 0, n_init = 'auto').fit(all_data)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
```

```python
plt.scatter(all_data[:, 0], all_data[:, 1], c=cluster_labels, cmap='viridis', marker='o', edgecolors='k')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=100, label='Cluster Centers')

# Add labels and legend
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KMeans Clustering')
plt.legend()

# Show the plot
plt.show()
```
