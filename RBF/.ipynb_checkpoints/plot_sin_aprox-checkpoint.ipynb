```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funct_sin(arr_ : np.ndarray):
    return np.sin(arr_)
```

```python
points = np.arange(start = -2*np.pi, stop = 2*np.pi, step = 0.5)
points_cont = np.linspace(start = -2*np.pi, stop = 2*np.pi, num = 300)
y = funct_sin(points)
y_cont = funct_sin(points_cont)
plt.scatter(points, y, color = 'black', label = 'func_generator fg(x)')
plt.plot(points_cont, y_cont, color = 'black')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
import numpy as np

def gaussian_function(arr_, r, m):
    return (1/np.sqrt(2*np.pi*r**2))*np.exp(-0.5* ((arr_ - m)**2) / (2*r**2))
```
