```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_generator1(arr_ : np.ndarray):
    return arr_ ** 2
def func_generator2(arr_ : np.ndarray):
    return (arr_ - 5) ** 2
```

```python
np.random.seed(123)
```

```python
x = np.linspace(start = - 5, stop = 10, num = 200)
```

```python
Y1 = func_generator1(x)
Y2 = func_generator2(x)
```

```python
np.random.seed(123)
plt.plot(x, Y1, color = 'blue')
plt.plot(x, Y2, color = 'red')
plt.xlim(-20, 20)
plt.ylim(0, 20)
plt.xlabel("w")
plt.ylabel("J")
plt.title('J x w')
```

```python
x1_scat = np.arange(start = - 5, stop = 10, step = 0.1)
Y1_scat = func_generator1(x1_scat)
Y2_scat = func_generator2(x1_scat)
```

```python
np.random.seed(123)
plt.scatter(x1_scat, Y1_scat, color = 'blue')
plt.scatter(x1_scat, Y2_scat, color = 'red')
plt.xlabel('w')
plt.ylabel('J')
plt.axvline(x = 0, color = 'black')
plt.axvline(x = 5, color = 'black')
plt.xlim(-20, 20)
plt.ylim(0, 20)
plt.title('J x w')
```

```python
region1 = (x1_scat <= 0)
region2 = (0 <= x1_scat) & (x1_scat <= 5)
region3 = (5 <= x1_scat)
```

```python
np.random.seed(123)
plt.scatter(Y1_scat[region1], Y2_scat[region1], color = 'blue')
plt.scatter(Y1_scat[region2], Y2_scat[region2], color = 'red')
plt.scatter(Y1_scat[region3], Y2_scat[region3], color = 'blue')
plt.title('Grfico de Pareto')
plt.axvline(x = 25, color = 'black')
plt.axvline(x = -1, color = 'black')
#plt.legend()
plt.xlabel("Y1")
plt.ylabel("Y2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funo do segundo grau (exemplo: f(x) = x^2)
def funcao_quadratica(x):
    return x**2

# Gerando valores de x e y para a funo original
x = np.linspace(-5, 5, 100)  # Valores de x
y = funcao_quadratica(x)      # Valores de y

# Deslocamento da funo no eixo x
deslocamento_x = 2
x_deslocado = x + deslocamento_x

# Plotando a funo original e a funo deslocada
plt.plot(x, y, label='Funo Original')
plt.plot(x_deslocado, y, label='Funo Deslocada')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Deslocamento de uma Funo Quadrtica no Eixo X')
plt.legend()
plt.grid(True)
plt.show()
```
