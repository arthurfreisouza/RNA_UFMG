```python
def sinc_func(arr_ ):
    return (np.sin(arr_) / arr_)
```

```python
np.random.seed(123)
x = np.linspace(start = - 10, stop = 10, num = 200)
y = sinc_func(x)
plt.plot(x, y, label = 'sinc function', color = 'blue')
plt.legend()
plt.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian_function(arr_, r, m):
    return (1/np.sqrt(2*np.pi*r**2))*np.exp(-0.5* ((arr_ - m)**2) / (2*r**2))
```

```python
xl = -10
xr = 10
st = 2
p = 10
faixa = (xr - xl) / p
c0 = xl + 0.5*faixa

x = np.arange(start = xl, stop = xr, step = st)
y = np.sinc(x)
```

```python
#y.shape
```

```python
plt.scatter(x, y, color = 'black', label = 'sinc(x)', s = 1)
plt.legend()
plt.plot()
```

```python
# Alocação dos centros.
ci = np.zeros(p) # p é o parametro que definirá os centros.
ci[0] = -9
for i in range(1, p):
    ci[i] = ci[i - 1] + faixa
```

```python
# Criando as funções radiais.
# Em cada coluna conterá 1 função radial.
H = np.zeros((x.shape[0], p))
for j in range(x.shape[0]):
    for i in range(p):
        H[: ,i] = gaussian_function(x, m = ci[i], r = 0.6)
```

```python
xrange = np.linspace(start = xl, stop = xr, num = 400)
```

```python
Hrange = np.zeros((xrange.shape[0], p))
for j in range(xrange.shape[0]):
    for i in range(p):
        Hrange[: ,i] = gaussian_function(xrange, m = ci[i], r = 0.6)
```

```python
ph = np.linalg.pinv(H)
w = np.dot(ph, y)
```

```python
yhat = np.dot(H, w)
yhatr = np.dot(Hrange, w)
```

```python
plt.plot(xrange, yhatr, color = 'blue')
plt.scatter(x, yhat, color = 'red')
plt.xlim(-30, 30)
plt.show()
```

```python
%%R
rm(list = ls())
library('corpcor')

generate_gaussian <- function(n, mean = 0, sd = 1) {
  # Generate n random numbers from a Gaussian distribution
  return(rnorm(n, mean = mean, sd = sd))
}

xl <- -10
xr <- 10
st <- 0.01
p <- 10
faixa <- (xr - xl) / p
c0 <- xl + 0.5 * faixa
ci <- rep(0, p)
ci[1] <- c0
for (i in 2:p)
  ci[i] <- ci[i - 1] + faixa

x <- seq(xl, xr, st)
y <- sin(x) / x

H <- matrix(nrow = length(x), ncol = p)
for (i in 1:p)
  H[, i] <- generate_gaussian(length(x), ci[i], 0.6)

ph <- pseudoinverse(H)
w <- ph %*% y
yhat <- H %*% w

xrange <- seq(xl, xr, 0.05)
Hrange <- matrix(nrow = length(xrange), ncol = p)
for (i in 1:p)
  Hrange[, i] <- generate_gaussian(length(xrange), ci[i], 0.6)

yhatr <- Hrange %*% w

# Plot original function
plot(x, y, type = 'l', xlim = c(xl, xr), ylim = c(-1, 1), xlab = "x", ylab = "f(x), h1(x), h2(x)")

# Plot approximated function
lines(xrange, yhatr, col = 'red')

# Plot individual basis functions
for (i in 1:p){
  plot(x, H[, i] + i/10, type = 'l', col = 'blue')  # Adjust the y-value to separate the lines vertically
}
```
