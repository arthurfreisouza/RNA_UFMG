```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("diabetes_prediction_dataset.csv")
df[df['diabetes'] == 0].value_counts()
```

```python
(df['diabetes'].value_counts(dropna = False) / df.shape[0])*100
df
```

```python
df_d = pd.get_dummies(df)
```

```python
df_d
```

```python
df_d['gender_Female'].value_counts(dropna = False)
```

```python
X = df_d.div(df_d.sum(axis = 1), axis = 'rows')
```

```python
X
```

```python
km = KMeans(n_clusters = 8)
km.fit_predict(X)
```

```python
distortions = []
n_clusters = range(2,15)
for n_clus in n_clusters:
    print(n_clus)
    distortions.append(KMeans(n_clusters = n_clus, max_iter = 10_000, n_init = 100, random_state = 61658).fit(X).inertia_)
```
