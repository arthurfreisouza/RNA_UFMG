```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

##### Gerando os dados a serem clusterizados.

```python
X = -2 * np.random.rand(100, 2) # Criando valores aleat√≥rios no formato desejado.
X1 = 1 + 2 * np.random.rand(50, 2)
X[50 : 100, :] = X1
plt.scatter(X[ :, 0], X[ :, 1], s = 50, c = 'b')
plt.show()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters = 2)
Kmean.fit(X)
print(Kmean)
```

```python
Kmean.cluster_centers_ # Da a coordenada dos centros dos clusters.
```

```python
plt.scatter(X[ :, 0], X[ :, 1], s = 50, c = 'b')
plt.scatter(-1.12654815, -0.91723407, s = 200, c = 'green', marker = 's')
plt.scatter(2.01559419, 2.02597093, s = 200, c = 'red', marker = 's')
plt.show()
```

```python
Kmean.labels_
```

```python
sample_test = np.array([-3.0, -3.0])
second_test = sample_test.reshape(1, -1)
Kmean.predict(second_test)
```
