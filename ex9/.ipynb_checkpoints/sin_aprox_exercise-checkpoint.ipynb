```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sech2(x):
    y = np.array(2 / (np.exp(x) + np.exp(-x))) * np.array(2 / (np.exp(x) + np.exp(-x)))
    return y
```

```python
x = np.arange(start = 0, stop = np.pi, step = 0.1)
x_cont = np.linspace(start = 0, stop = 2*np.pi, num = 65)
y = np.sin(x) + np.random.uniform(low = -0.1, high = 0.1, size = x.shape[0])
```

```python
plt.scatter(x, y, label = 'sin(x) + gaussian noise', color = 'red', s = 5)
plt.plot(x_cont, np.sin(x_cont), label = 'sin(x)', color = 'black')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
w111 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
w112 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
w113 = np.random.uniform(low = -0.5, high = 0.5, size = 1)

wb111 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wb112 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
wb113 = np.random.uniform(low = -0.5, high = 0.5, size = 1)

w124 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
w224 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
w324 = np.random.uniform(low = -0.5, high = 0.5, size = 1)

wb224 = np.random.uniform(low = -0.5, high = 0.5, size = 1)
```

```python
bias1 = 1
bias2 = 1
```

```python
num_epochs = 0
tol = 0.01
err_epochs = tol + 1
max_epochs = 1500
N = x.shape[0] # Número de entradas.
eta = 0.1 # Learning_rate.
err_vec = np.zeros(max_epochs, dtype=float, order = 'C')
```

```python
sequence = np.arange(start = 0, stop = N, step = 1).tolist()
sequence
seq_randomized = random.sample(population = sequence, k = N - 1)
seq_randomized
```

```python
while((num_epochs < max_epochs) and (err_epochs > tol)):
    sequence = np.arange(start = 0, stop = N, step = 1).tolist()
    seq_randomized = random.sample(population = sequence, k = N)
    err2 = 0 # Erro de cada época.

    for i in range(N):
        rand_input = seq_randomized[i]

        ipt = x[rand_input] # Entrada aleatória
        opt = y[rand_input] # Saída aleatória


        # Combinação linear 1 camada: 

        cl1 = (bias1*wb111) + (ipt*w111)
        tan_cl1 = np.tanh(cl1)


        cl2 = (bias1*wb112) + (ipt*w112)
        tan_cl2 = np.tanh(cl2)


        cl3 = (bias1*wb113) + (ipt*w113)
        tan_cl3 = np.tanh(cl3)




        # Saída da rede : 

        opt_net = (tan_cl1*w124) + (tan_cl2*w224) + (tan_cl3*w324) + (bias2*wb224)
        err_opt = (opt - opt_net)
        delta_opt = err_opt*1

        delta_bias2 = (eta*delta_opt*bias2)
        delta_w124 = (eta*delta_opt*cl1)
        delta_w224 = (eta*delta_opt*cl2)
        delta_w324 = (eta*delta_opt*cl3)


        delta_ipt21 = (delta_opt*w124)*sech2(cl1)
        delta_ipt22 = (delta_opt*w224)*sech2(cl2)
        delta_ipt23 = (delta_opt*w324)*sech2(cl3)



        dbias11 = (eta*delta_ipt21*bias1)
        dbias12 = (eta*delta_ipt22*bias1)
        dbias13 = (eta*delta_ipt23*bias1)


        dipt11 = (eta*delta_ipt21*ipt)
        dipt12 = (eta*delta_ipt22*ipt)
        dipt13 = (eta*delta_ipt23*ipt)

        wb111 = wb111 + dbias11
        wb112 = wb112 + dbias12
        wb113 = wb113 + dbias13

        
        # Atualizando pesos camada de entrada.
        w111 = w111 + dipt11
        w112 = w112 + dipt12
        w113 = w113 + dipt13


        # Atualizando pesos camada de saída.
        wb224 = wb224 + delta_bias2
        w124 = w124 + delta_w124
        w224 = w224 + delta_w224
        w324 = w324 + delta_w324

        err2 = err2 + (err_opt**2)

    
    err_vec[num_epochs] = err2/N
    err_epochs = err_vec[num_epochs]
    num_epochs += 1

        
        
```

```python
err_vec.shape
```

```python
x_plot = np.linspace(start = 0, stop = num_epochs, num = err_vec.shape[0])
plt.plot(x_plot, err_vec, color = 'black', label = 'err_graphic')
plt.title('Err graphic')
plt.xlabel('num_epochs')
plt.ylabel('err²')
plt.legend()
plt.show()
```

```python
xrange = np.linspace(start = 0, stop = np.pi, num = x.shape[0]*100)
yhat = np.zeros((xrange.shape[0]))
N2 = xrange.shape[0]
```

```python
for i in range(N2):
    ipt_test = xrange[i]

    cbl1 = (bias1*wb111) + (ipt_test*w111)
    otp_n1_hidd   = np.tanh(cbl1)

    cbl2 = (bias1*wb112) + (ipt_test*w112)
    otp_n2_hidd   = np.tanh(cbl2)
    
    cbl3 = (bias1*wb113) + (ipt_test*w113)
    otp_n3_hidd   = np.tanh(cbl3)

    final_otp = (bias2*wb224) + (otp_n1_hidd*w124) + (otp_n2_hidd*w224) + (otp_n3_hidd*w324)

    yhat[i] = final_otp    
```

```python
plt.plot(xrange, yhat, color = 'red', label = 'sin(x) aproxximation')
#plt.plot(xrange, np.sin(xrange) + 0.25, color = 'blue')
plt.plot(xrange, np.sin(xrange), color = 'black', label = 'sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(np.sin(xrange), yhat)
```
