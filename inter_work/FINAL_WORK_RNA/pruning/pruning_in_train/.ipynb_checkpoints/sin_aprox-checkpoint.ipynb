```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import techniques
```

```python
x = np.arange(start = -7, stop = 7, step = 0.01) 
y = 10*x**2 + 200*np.random.normal(loc = 0, scale = 0.5, size = x.shape[0])
```

```python
plt.scatter(x, y, color = 'black')
plt.title("sin(x)")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
p_n = 5
keep_val = np.arange(start = 0.25, stop = 0.99, step = 0.1)
best_keep, lowest_MSE, arr_ret = techniques.grid_searchCV_pruning(xin = x, yin = y, p = p_n, keep_rate = keep_val, CV_groups = 10, classification = False)
```

```python
x_plot = np.arange(start = 0.25, stop = keep_val[keep_val.shape[0] -1] + 0.1, step = 0.1)
plt.plot(x_plot, arr_ret, color = 'black', label = 'MSE vs keep_rate')
plt.title('MSE vs keep_rate')
plt.xlabel('keep_rate')
plt.ylabel('MSE')
plt.legend()
plt.show()
```

```python
ret = techniques.train_ELM_PRUNING(xin = x, yin = y, p = p_n, control = True, keep_rate = best_keep)
best_w = ret[0]
best_h = ret[1]
best_z = ret[2]
err_vec = ret[3]
```

```python
x_plot = np.arange(start = 0, stop = err_vec.shape[0], step = 1)
plt.plot(x_plot, err_vec, color = 'black', label = 'error vs epoch')
plt.title('error vs epoch')
plt.xlabel('epoch')
plt.ylabel('error')
plt.legend()
plt.show()
```

```python
xrange = np.linspace(start = -7, stop = 7, num = 50)
y_net = techniques.test_ELM(xin = xrange, Z = best_z, W = best_w, control = True, classification = False)
```

```python
plt.scatter(xrange, y_net, color = 'black', label = 'aproximation')
plt.scatter(xrange, 10*xrange**2 , color = 'red', label = '10*sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Aproximation')
plt.legend()
plt.show()
```
