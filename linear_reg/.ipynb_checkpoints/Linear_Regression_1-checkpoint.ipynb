```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import operator
```

```python
np.random.seed(0)
x = 2 - 3 * np.random.normal(loc = 0, scale = 1, size = 100)
y = x - 5 * (x**2) + (x**3)


x = x[:, np.newaxis]
y = y[:, np.newaxis]

plt.scatter(x, y, color = 'black', s = 3)
```

```python
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
plt.scatter(x, y, color = 'black', s = 3)
plt.plot(x, y_pred, color = 'red')
```

```python
polinomial_ped = PolynomialFeatures(degree = 3)
x_pol = polinomial_ped.fit_transform(x)
model2 = LinearRegression()
model2.fit(x_pol, y)
y_pol_pred = model2.predict(x_pol)

rmse = np.sqrt(mean_squared_error(y, y_pol_pred))
r2 = r2_score(y, y_pol_pred)
print(f"rmse : {rmse}")
print(f"r2_score : {r2}")
```

```python
plt.scatter(x, y, color = 'black', s = 3)
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x, y_pred), key = sort_axis)
x, y_pol_pred = zip(*sorted_zip)
plt.plot(x, y_pol_pred, color = 'red')
plt.show()
```
