```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt 
```

```python
# Input features
X = np.array([1, 5, 2, 8, 4, 5]).reshape(-1, 1)
# Output variable
y = np.array([2, 4, 5, 4, 5, 6.9])
```

```python
degree = 3
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
```

```python
# Create a linear regression object
model = LinearRegression()
# Train the model using the polynomial features
model.fit(X_poly, y)
```

```python
# Generate polynomial features for prediction
X_test = np.linspace(start = 1, stop = 10, num = 100).reshape(-1, 1) #numpy.linspace(start,stop,num= 50)

X_test_poly = poly_features.transform(X_test)

# Predict the output for new data points
y_pred = model.predict(X_test_poly)
```

```python

# Plot the original data points
plt.scatter(X, y, color='b', label='Actual')
# Plot the predicted values
plt.plot(X_test, y_pred, color='r', label='Predicted')
# Add labels and title to the plot
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression (Degree {})'.format(degree))
plt.legend()
# Display the plot
plt.show()
```
