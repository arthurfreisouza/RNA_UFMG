```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando a função que irá gerar os dados.
def func_generator(arr_ : np.ndarray):
    return (0.5 * (arr_ ** 2) + 3 * arr_ + 10)
```

```python
# Gerando os dados, juntamente com 1 ruído gaussiano de desvio padrão 4.
np.random.seed(123)
X = np.random.uniform(low = -15, high = 10, size = 20)
Y = func_generator(X) + 4 * np.random.randn(len(X))
plt.scatter(X, Y, color = 'blue')
plt.title('Dados gerados com ruído')
```

```python
# A função H é a entrada dos neurônios da camada de saída. Ela resulta nos valores da entrada
# x elevados a uma certa potência. A última coluna é uma coluna apenas de valores 1s.
# w é a pseudoinversa de H mu
H = np.column_stack([ X**2, X, np.ones_like(X)])
w = np.linalg.pinv(H) @ Y 
```

```python
# Criando e plotando os gráficos contínuos para criar 1 representação gráfica.
xgrid = np.linspace(-15, 10, 1000) # Gera 1 distribuição de 1000 pontos que começa em -15 até o 10.
ygrid = func_generator(xgrid) # é a função, aplicada nos valores continuos xgrid
Hgrid = np.column_stack([xgrid**2, xgrid, np.ones_like(xgrid)]) 
yhatgrid = Hgrid @ w # Calculando os valores do y contínuos (y = Hw)
```

```python
plt.plot(xgrid, ygrid, color='blue', label='Função Original') # Plotando o gráfico da função.
plt.scatter(X, Y, color = 'red', label = 'Dados que foram originados da função') # Dados
plt.plot(xgrid, yhatgrid, color = 'green', label = 'Função aproximadora')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Aproximação de Função')
plt.legend()
plt.show()
```
